抓Glitch上DCard的資料

```python
import requests
import json
import pandas as pd
import jieba

newWords = ["沒安全感 1000", "金剛超人 30", "NewJeans 30", "前情提要 100"]
f = open("dict.txt", "w")
f.write("\n".join(newWords))
f.close()

jieba.load_userdict("dict.txt") #結巴字典

keywords= ["五月天", "blackpink", "bts", "胡瓜", "大谷翔平"]
url = "https://sincere-positive-saturnalia.glitch.me/dcard/{keyword}"

emotion = {} #紀錄每一個關鍵字的正面次數、負面次數、以及正面加總，與負面加總
for word in keywords:
  data = requests.get(url.format(keyword=word))
  print(word)
  result = json.loads(data.text)

  temp = []
  commons = []
  for item in result:
    if item["title"] is not None:
      text = item["title"]
      text = list(jieba.cut(text))
      text = " ".join(text)[0:512]
      result = pipeline('sentiment-analysis')(text)
      if result[0]["label"] == "POSITIVE":
        print(result)
        print(text)
      if word in emotion:
        if result[0]["label"] == "POSITIVE":
          emotion[word]["positive"] += 1
          emotion[word]["positiveSum"] += result[0]["score"]
        else:
          emotion[word]["negative"] += 1
          emotion[word]["negativeSum"] += result[0]["score"]
      else:
        if result[0]["label"] == "POSITIVE":
          emotion[word] = {"positive": 1, "negative": 0, "positiveSum": result[0]["score"], "negativeSum": 0}
        else:
          emotion[word] = {"positive": 0, "negative": 1, "positiveSum": 0, "negativeSum": result[0]["score"]}

print(emotion)
```
